```python
import pandas as pd
```

```python
pd.options.display.max_rows = 100
```

```python
artistdata = pd.read_csv('artist_data.csv', delimiter = ',', encoding= 'utf-8')
artistdata = artistdata.drop(columns = 'url')
artworkdata = pd.read_csv('artwork_data.csv',delimiter = ',', encoding= 'utf-8')
artworkdata = artworkdata.drop(columns = 'thumbnailUrl')
artworkdata = artworkdata.drop(columns = 'url')
artworkdata = artworkdata.drop(columns = 'thumbnailCopyright')
```

```python
artistdata = artistdata.rename(columns={'id': 'artistId'})
artworkdata = artworkdata.rename(columns={'id': 'artworkId'})
df = pd.merge(artistdata, artworkdata, on='artistId')
df = df.drop(columns = 'dates')
df = df.drop(columns = 'artist')
df
```

```python
#What kind of materials do women prefer?
women_filter = df['gender'] == 'Female'
womenartists = df[women_filter]

womenartists['medium'].value_counts()
```

```python
womenartists['medium'].value_counts()\
[:10].plot(kind= 'barh', title= 'most frequent medium used by women')
```

```python
#In comparison, what kind of materials do men prefer?
men_filter = df['gender'] == 'Male'
menartists = df[men_filter]

menartists['medium'].value_counts()\
[:10].plot(kind= 'barh', title= 'most frequent medium used by men')
```

```python
# We see that men have a big preference for graphite on paper,
# Whereas women have a smaller preference for Lithograph on paper.
# But we do need to take into account that there are more male artists than female artist!!
```
