```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get("https://www.azlyrics.com/lyrics/metallica/lux-aeterna.html")
html_string = response.text
```

```python
document = BeautifulSoup(html_string, "html.parser")
document
```

```python
song_title = document.find("h1").text
song_title = song_title[1:12]
print(song_title)
```

```python
#song_lyrics = document.find_all("div", attrs = {'class': None})
result = [x.get_text() for x in document.find_all("div", attrs = {'class': None})]
print(result)
```

```python
song_lyrics = ''.join(map(str,result))
print(song_lyrics)
```

```python
artist = document.find("h2").text
artist = artist[0:10]
print(artist)
```
